```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='gender', hue='Churn', data=customer_data)
plt.title("Churn by Gender")
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=customer_data)
plt.title("Churn by Contract Type")
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=customer_data)
plt.title("Monthly Charges vs Churn")
plt.show()
```

🔍 Insights:

1. A significant number of customers have churned compared to retained customers.
2. Customers with month-to-month contracts show higher churn rates.
3. Customers with higher monthly charges are more likely to churn.
4. Long-term contract customers (1 year / 2 year) show better retention.
5. Gender does not have a major impact on customer churn.


💡 Recommendations:

1. Encourage customers to switch to long-term contracts by offering discounts.
2. Reduce pricing or provide offers for customers with high monthly charges.
3. Improve customer engagement during initial months of subscription.
4. Introduce loyalty programs to retain existing customers.
5. Provide better customer support to reduce dissatisfaction.
